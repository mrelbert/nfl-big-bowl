## Import files for review

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

players = pd.read_csv('./players.csv')
games = pd.read_csv('./games.csv')
plays = pd.read_csv('./plays.csv')
week1 = pd.read_csv('./week1.csv')
```

## Merge week 1 with additional players data

```python
week1.shape
```

```python
# merged keys to the left on column 'nflId'

week_plus_addtl_player_info = pd.merge(week1, players, how='left', on='nflId')
week_plus_addtl_player_info.shape
```

```python
week_plus_addtl_player_info.head()
```

```python
week_player_games = pd.merge(week_plus_addtl_player_info, games, how='left', on='gameId')
week_player_games.shape
```

```python
week_player_games.head()
```

```python
week_player_games_play = pd.merge(week_player_games, plays, how='left', on=['gameId', 'playId'])
week_player_games_play.shape
```
